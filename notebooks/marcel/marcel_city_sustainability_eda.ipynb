```python
import os
import numpy as np
import pandas as pd
from PIL import Image

def get_label_value_counts(data_path):
    # Get label images path
    all_labels_path = []

    # Iterate over the subdirectories
    for subdir in os.listdir(data_path):
        subdir_path = os.path.join(data_path, subdir)
        if os.path.isdir(subdir_path):
            labels_dir = os.path.join(subdir_path, 'labels')

            # Get the list of image files in the "labels" directory
            image_files = [os.path.join(labels_dir, file) for file in os.listdir(labels_dir) if file.endswith('.tif')]

            # Add image files to the all_labels_path list
            all_labels_path += image_files

    # Numpy Array of all label images
    all_labels_np = []

    # Iterate over the image file paths
    for label_path in all_labels_path:
        lb_1 = Image.open(label_path)
        numpy_array_label = np.array(lb_1)
        lb_1.close()

        # Add the NumPy array to the all_labels_np list
        all_labels_np.append(numpy_array_label)

    # From the np arrays, get the count values for each class
    class_value_counts = []

    # Iterate over the NumPy arrays
    for label_np in all_labels_np:
        unique_classes, class_counts = np.unique(label_np, return_counts=True)
        value_counts_dict = dict(zip(unique_classes, class_counts))
        class_value_counts.append(value_counts_dict)

    # Merge the value counts for all NumPy arrays
    merged_value_counts = {}
    for value_counts in class_value_counts:
        for cls, count in value_counts.items():
            merged_value_counts[cls] = merged_value_counts.get(cls, 0) + count

    # Sort the merged value counts in ascending order
    sorted_value_counts = dict(sorted(merged_value_counts.items()))

    # From the merged value counts, get the count values for each class in percentage
    total_count = sum(sorted_value_counts.values())

    # Calculate the percentage value counts
    percentage_value_counts = {
        cls: count / total_count * 100 for cls, count in sorted_value_counts.items()}

    # Sort the percentage value counts in ascending order by class
    sorted_percentage_value_counts = {
        cls: percentage_value_counts[cls] for cls in sorted(percentage_value_counts)}

    return sorted_value_counts, sorted_percentage_value_counts

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with your main directory path

# Call the function to get the value counts and percentage value counts
value_counts, percentage_value_counts = get_label_value_counts(data_path)

# Class names mapping
class_names = {
    0: 'Other',
    1: 'Bareland',
    2: 'Rangeland',
    3: 'Developed space',
    4: 'Road',
    5: 'Tree',
    6: 'Water',
    7: 'Agriculture land',
    8: 'Building'
}

# Create a DataFrame for value counts
df_value_counts = pd.DataFrame.from_dict(value_counts, orient='index', columns=['Count'])
df_value_counts.index = df_value_counts.index.map(class_names)

# Create a DataFrame for percentage value counts
df_percentage_value_counts = pd.DataFrame.from_dict(percentage_value_counts, orient='index', columns=['Percentage'])
df_percentage_value_counts.index = df_percentage_value_counts.index.map(class_names)

# Print the DataFrames
print('Label count values')
print(df_value_counts)
print('\n')

print('Label count percentage')
print(df_percentage_value_counts)
```

##  def find_empty_folders(data_path)

```python
import os

def find_empty_folders(data_path):
    empty_image_folders = []
    empty_label_folders = []

    # Iterate over the subdirectories
    for subdir in os.listdir(data_path):
        subdir_path = os.path.join(data_path, subdir)
        if os.path.isdir(subdir_path):
            images_dir = os.path.join(subdir_path, 'images')
            labels_dir = os.path.join(subdir_path, 'labels')

            # Check if the images directory is empty
            if len(os.listdir(images_dir)) == 0:
                empty_image_folders.append(images_dir)

            # Check if the labels directory is empty
            if len(os.listdir(labels_dir)) == 0:
                empty_label_folders.append(labels_dir)

    return empty_image_folders, empty_label_folders

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with the actual directory path

# Call the function to find the empty image and label folders
empty_image_folders, empty_label_folders = find_empty_folders(data_path)

# Print the empty image folders
print("Empty image folders:")
for folder in empty_image_folders:
    print(folder)

print("\nEmpty label folders:")
for folder in empty_label_folders:
    print(folder)
```

## def folder_stats(data_path)

```python
import os

def get_folder_stats(data_path):
    total_images = 0
    total_labels = 0
    total_images_with_labels = 0

    # Iterate over the subdirectories
    for subdir in os.listdir(data_path):
        subdir_path = os.path.join(data_path, subdir)
        if os.path.isdir(subdir_path):
            images_dir = os.path.join(subdir_path, 'images')
            labels_dir = os.path.join(subdir_path, 'labels')

            # Count the number of images
            total_images += len(os.listdir(images_dir))

            # Count the number of labels
            total_labels += len(os.listdir(labels_dir))

            # Check if the images directory has corresponding labels
            image_files = os.listdir(images_dir)
            label_files = os.listdir(labels_dir)
            images_with_labels = set(image_files).intersection(label_files)
            total_images_with_labels += len(images_with_labels)

    return total_images, total_labels, total_images_with_labels

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with the actual directory path

# Call the function to get the folder statistics
total_images, total_labels, total_images_with_labels = get_folder_stats(data_path)

# Print the results
print(f"Total number of images: {total_images}")
print(f"Total number of labels: {total_labels}")
print(f"Total number of images with labels: {total_images_with_labels}")
```

## def count_images_by_size(folder_path)

```python
import os
from collections import Counter
from PIL import Image

def get_image_dimensions(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
    return width, height

def count_images_by_size(data_path):
    image_sizes = Counter()

    # Iterate over the subdirectories in the main data path
    for subdir in os.listdir(data_path):
        subdir_path = os.path.join(data_path, subdir)
        if os.path.isdir(subdir_path):
            images_folder = os.path.join(subdir_path, 'images')

            # Iterate over the image files in the images folder
            for file_name in os.listdir(images_folder):
                file_path = os.path.join(images_folder, file_name)
                if os.path.isfile(file_path):
                    width, height = get_image_dimensions(file_path)
                    image_sizes[(width, height)] += 1

    return image_sizes

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with the actual directory path

# Count images by size in the images folder present in each subfolder
image_sizes = count_images_by_size(data_path)

# Print the counts per different sizes in the images folder
print("Images folder - Count per different sizes:")
for size, count in image_sizes.items():
    print(f"Size: {size} - Count: {count}")
```

```python
def get_label_dimensions(label_path):
    with Image.open(label_path) as img:
        width, height = img.size
    return width, height

def count_labels_by_size(data_path):
    label_sizes = Counter()

    # Iterate over the subdirectories in the main data path
    for subdir in os.listdir(data_path):
        subdir_path = os.path.join(data_path, subdir)
        if os.path.isdir(subdir_path):
            labels_folder = os.path.join(subdir_path, 'labels')

            # Iterate over the label files in the labels folder
            for file_name in os.listdir(labels_folder):
                file_path = os.path.join(labels_folder, file_name)
                if os.path.isfile(file_path):
                    width, height = get_label_dimensions(file_path)
                    label_sizes[(width, height)] += 1

    return label_sizes

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with the actual directory path

# Count labels by size in the labels folder present in each subfolder
label_sizes = count_labels_by_size(data_path)

# Print the counts per different sizes in the labels folder
print("Labels folder - Count per different sizes:")
for size, count in label_sizes.items():
    print(f"Size: {size} - Count: {count}")
```

## def process_random_image(data_path)

```python
import os
import random
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the class colors dictionary
class_colors = {
    0: '#FFFF00',   # Class 0 - Yellow
    1: '#800000',   # Class 1 - Maroon
    2: '#00FF24',   # Class 2 - Lime Green
    3: '#949494',   # Class 3 - Grey
    4: '#FFFFFF',   # Class 4 - White
    5: '#226126',   # Class 5 - Forest Green
    6: '#0045FF',   # Class 6 - Blue
    7: '#4BB549',   # Class 7 - Dark Green
    8: '#DE1F07'}   # Class 8 - Red


# Define the class descriptions dictionary
class_descriptions = {
    0: 'Other',
    1: 'Bareland',
    2: 'Rangeland',
    3: 'Developed Space',
    4: 'Road',
    5: 'Tree',
    6: 'Water',
    7: 'Agriculture Land',
    8: 'Building'}

def process_random_image(data_path):
    # Get a random subdirectory
    random_subdir = random.choice(os.listdir(data_path))
    subdir_path = os.path.join(data_path, random_subdir)

    # Get a random image file within the labels folder of the subdirectory
    labels_dir = os.path.join(subdir_path, 'labels')
    random_image_file = random.choice(os.listdir(labels_dir))
    file_path = os.path.join(labels_dir, random_image_file)

    # Open the image and convert it to a numpy array
    lb_1 = Image.open(file_path)
    numpy_array_label = np.array(lb_1)

    # Perform value counts on the numpy array
    value_counts = pd.DataFrame(numpy_array_label.reshape(-1, 1)).value_counts().sort_index()

    # Create a color-mapped version of the image
    color_mapped_image = np.zeros((lb_1.height, lb_1.width, 3), dtype=np.uint8)
    for cls, color in class_colors.items():
        color_indices = (numpy_array_label == cls)
        color_mapped_image[color_indices] = tuple(int(color.lstrip('#')[i:i+2], 16) for i in (0, 2, 4))
    
    # Plot the images and value counts
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
    
    # Display the original image
    ax1.imshow(lb_1)
    ax1.axis('off')
    ax1.set_title(f"Original Image: {random_image_file}")
    
    # Display the color-mapped image with class labels outside the image
    ax2.imshow(color_mapped_image)
    ax2.axis('off')
    ax2.set_title("Color-Mapped Image")

    # Add class labels outside the image
    for cls, color in class_colors.items():
        description = class_descriptions[cls]
        ax2.text(0, -10 * cls, f"{description}", color=color, fontsize=10, fontweight='bold')

    # Convert value counts to a DataFrame with class descriptions
    value_counts_df = pd.DataFrame({'Class': value_counts.index.get_level_values(0),
                                    'Count': value_counts.values})
    value_counts_df['Description'] = value_counts_df['Class'].map(class_descriptions)
    value_counts_df = value_counts_df.sort_values('Class').reset_index(drop=True)
    
    # Display the value counts as a DataFrame
    print("Value Counts:")
    print(value_counts_df)
    # Show the plot
    plt.tight_layout()
    plt.show()

# Define the directory path
data_path = 'raw_data/OpenEarthMap_wo_xBD'  # Replace with the actual directory path

# Call the function to process a random image and display the value counts with class colors

process_random_image(data_path)
```
